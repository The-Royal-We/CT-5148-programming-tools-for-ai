### Computational complexity

What is the time complexity of the following algorithm?

````python
def binary_search(L, value):
    '''
    from https://rosettacode.org/wiki/Binary_search#Python
    
    >>> L = [1, 2, 3, 5, 8, 13, 21, 34] # sorted, no duplicates
    >>> binary_search(L, 1)
    0
    >>> binary_search(L, 36)
    -1
    >>> binary_search(L, 5)
    3
    >>> binary_search(L, -1)
    -1
    >>> binary_search(L, 9)
    -1
    '''
    low = 0
    high = len(L) - 1
    while low <= high: 
        mid = (low + high) // 2
        if L[mid] > value: high = mid - 1
        elif L[mid] < value: low = mid + 1
        else: return mid
    return -1
````

**Exercise**: what is the computational complexity of the following algorithm?

````python
def subset(S, T):
    '''
    Return True if S is a subset of T, ie every element of S is in T also. 
    Both S and T will be lists. T might not be sorted, so we sort it.
    '''
    T.sort()
    return all((binary_search(T, x) >= 0) for x in S)
````
